### Data Visualization With Matplotlib
Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]


#creting a linePlot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title("Basic Line Plot")
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]


#customized line plot
plt.plot(x,y,color='red',linestyle='--')
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]


#customized line plot
plt.plot(x,y,color='red',linestyle='-.')
```

```python
#multiple plots
##sample data
x=[1,2,3,4,5]
y1=[i*i for i in x]
y2=[i for i in x]


plt.figure(figsize=(9,5)) #first value =width second height

plt.subplot(1,2,1)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 1 as first position
plt.plot(x,y1,color="red")
plt.title("plot 1")


plt.subplot(1,2,2)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 2 as first position
plt.plot(x,y1,color="green")
plt.title("plot 2")
```

```python
#multiple plots
##sample data
x=[1,2,3,4,5]
y1=[i*i for i in x]
y2=[i for i in x]


plt.figure(figsize=(9,5))

plt.subplot(2,2,1)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 1 as first position
plt.plot(x,y1,color="red")
plt.title("plot 1")


plt.subplot(2,2,2)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 2 as first position
plt.plot(x,y2,color="green")
plt.title("plot 2")

plt.subplot(2,2,3)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 2 as first position
plt.plot(y2,y1,color="blue")
plt.title("plot 3")

plt.subplot(2,2,4)#it specify 1 as number of row 2 as column(mean in 1 row 2 colum) and 2 as first position
plt.plot(y1,y2,color="pink")
plt.title("plot 4")
```

```python
#bar graph  plotting
category=['A','B','C','D','E']
values=[16,2,10,9,15]
plt.bar(category,values,color='red')
plt.title("Bar graph")
plt.xlabel('Category')
plt.ylabel('values')
```

```python
#bar graph  plotting
category=['A','B','C','D','E','A','B']
values=[16,2,10,9,15,25,2]
plt.bar(category,values,color='red')
plt.title("Bar graph")
plt.xlabel('Category')
plt.ylabel('values')
```


Histogram:-histogram are used to present the distribution of a dataset, They divide the data into bins and count the number of data points in each bin

```python
#sample data
import random
import numpy as np
data= np.random.randint(1,6,(20))
print(data)
plt.hist(data,bins=5,color='black',edgecolor='red')
```

# pie chart

```python
category=['A','B','C','D','E']
values=[16,2,10,9,15]
plt.pie(values,labels=category,autopct='%1.1f%%')
```

```python
#we can explicitly define color
category=['A','B','C','D','E']
values=[16,2,10,9,15]
color=['gold','orange','pink','red','gray']
explode=(0.2,0,0,0,0)
plt.pie(values,labels=category,colors=color,explode=explode,autopct='%1.1f%%',shadow=True)
```

```python
#sales data visualization
import pandas as pd
df=pd.read_csv('sales_data.csv')

df.head(5)
total_sales_by_product=df.groupby('Product Category')['Total Revenue'].sum()
print(total_sales_by_product)

```

```python
total_sales_by_product.plot(kind='bar',color='teal')
plt.ylabel("Total revenue")
```

```python
#Sales trends over time
trend_date_wise=df.groupby('Date')['Total Revenue'].sum().reset_index()
print(trend_date_wise)
plt.plot(trend_date_wise['Date'],trend_date_wise['Total Revenue'])
plt.xlabel('Date')
plt.ylabel('Total revenue')

```
